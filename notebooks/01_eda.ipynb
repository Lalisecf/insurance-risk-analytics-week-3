### Loss Ratio Formula

The Loss Ratio was calculated using the following formula:

```python
Loss Ratio = TotalClaims / TotalPremium
```

Loss Ratio is one of the most important insurance profitability metrics:

- A value greater than `1` indicates that claims exceed premiums collected (unprofitable segment).
- A value less than `1` indicates that premiums collected exceed claims paid (profitable segment).

---

### Interpretation

The analysis of loss ratio across vehicle types helps identify which categories contribute most to insurance risk.

### Key Findings

- Vehicle categories with higher loss ratios generate claims that are closer to or exceed collected premiums.

- Commercial and specialized vehicle categories may show different profitability patterns due to operational risk exposure.

### Business Insight

- High-risk vehicle categories may require:
  - premium adjustments,

  - or differentiated policy pricing.
- Low-risk vehicle categories could be targeted for customer acquisition campaigns through reduced premiums.

### Business Insight

- This is partially desirable from a profitability standpoint because it suggests that premiums are not perfectly driven by claim costs alone.
- However, the weak relationship may also indicate pricing inefficiencies in some market segments.

### Link to Previous Analysis

- This finding supports the earlier loss ratio analysis:
  - Provinces such as **Gauteng** displayed high loss ratios.
  - Provinces such as **Northern Cape** displayed low loss ratios.

This means profitability varies significantly across segments despite similar premium ranges.

### 2. TotalPremium vs CustomValueEstimate (0.02)

### Possible Implications

- Premium pricing may not adequately reflect asset value.

### Business Insight

- Vehicle or asset value may not currently be a major driver of premium pricing.
- This could expose ACIS to underpriced high-value risks.

### 3. TotalPremium vs SumInsured (-0.041)

- The very weak negative correlation is effectively close to zero.
- There is no meaningful linear relationship between premium amounts and insured sums.

### Business Insight

- The result suggests possible pricing inconsistencies or segmentation effects.

### Potential Concern

- Policies with large coverage amounts but relatively low premiums may be financially risky for ACIS.

### 4. TotalClaims vs CustomValueEstimate / SumInsured

- Both relationships are essentially zero.

### Interpretation

### Business Insight

Claims may instead be influenced more heavily by:
- geographic factors,
- vehicle type,
- driver behaviour,
- fraud risk,
- or policy characteristics.

## Overall Interpretation

The correlation matrix reveals that the major financial variables are largely independent in a linear sense.

### Key Observations

- Most correlations are extremely weak or negligible.
- There is no strong evidence of multicollinearity among these features.
- The dataset likely contains:
  - non-linear relationships,
  - segment-specific behaviour,
  - and hidden interactions between variables.

## Key Business Insights

| Finding | Business Implication |
